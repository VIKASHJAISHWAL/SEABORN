```python
sns.displot(data=titanic,x='age',kind='hist',element='step',hue='sex')
```

```python
#kdeplot
#Rather than using dicrete bins,a kde plot smooth the obeservationwith a gaussion keranel,producing a continus density estimate.
sns.kdeplot(data=tips,x='total_bill')

```

```python
sns.displot(data=tips,x='total_bill',kind='kde')
```

```python
#hue->fill
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex')
```

```python
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True)
```

```python
#Rugplot
#plot marginal distribution by drawing ticks along the x and y axes.
#this function is inteneded to complement other plots by showing  the location of individual oberservation in an obtrusive  way.
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

```python
sns.rugplot(data=tips,x='total_bill')
```

```python
# Bi variate histogram 
#bi variate histogrambins the data within each ractangle that tile the plot
#and then show the count of the observations with in each rectangle with the fill color
sns.histplot(data=tips,x="total_bill",y="tip")
```

```python
sns.displot(data=tips,x="total_bill",y="tip",kind="hist")
```

```python
#bivariate Kdeplot
#a bivariate KDE plot  Smoothes the(x,y) observations with a 2D Gaussian.
sns.kdeplot(data=tips,x="total_bill",y="tip")
```

```python
#HeatMap
#plot rectangular data as color-encoded matrix
gap
```

```python
gap.pivot(index='country',columns="year",values="lifeExp")
```

```python
temp_df=gap.pivot(index='country',columns="year",values="lifeExp")
#axes level function
plt.figure(figsize=(15,10))
sns.heatmap(temp_df)
```

```python
#annot
#hamko uropin company ka data chahiye
temp_df=gap[gap["continent"]=="Europe"].pivot(index='country',columns="year",values="lifeExp")
plt.figure(figsize=(15,10))
sns.heatmap(temp_df,annot=True)
```

```python
#annot
#hamko uropin company ka data chahiye
temp_df=gap[gap["continent"]=="Europe"].pivot(index='country',columns="year",values="lifeExp")
plt.figure(figsize=(15,10))
sns.heatmap(temp_df,annot=True,linewidth=0.5)
```

```python
#annot
#hamko uropin company ka data chahiye
temp_df=gap[gap["continent"]=="Europe"].pivot(index='country',columns="year",values="lifeExp")
plt.figure(figsize=(15,10))
sns.heatmap(temp_df,annot=True,linewidth=0.5,cmap="viridis")
```
